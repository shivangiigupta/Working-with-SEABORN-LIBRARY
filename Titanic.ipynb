```python
df = sns.load_dataset("titanic")
sns.boxplot(x=df["age"])
```

```python
sns.boxplot(data=df, x="age", y="class", hue="alive")
```

```python
sns.boxplot(data=df, x="fare", y="alive", order=["yes", "no"])
```

```python
sns.boxplot(data=df[["age", "fare"]], orient="h")
```

```python
sns.boxplot(data=df, x="fare", y="deck", hue="deck", dodge=False)
```

```python
sns.boxplot(data=df, x="age", y="class",notch=True, showcaps=False,flierprops={"marker": "x"},boxprops={"facecolor": (.4, .6, .8, .5)},medianprops={"color": "coral"},)
```

```python
df = sns.load_dataset("titanic")
sns.catplot(data=df, x="age", y="class")
```

```python
sns.catplot(data=df, x="age", y="class", kind="box")
```

```python
sns.catplot(data=df, x="age", y="class", hue="sex", kind="boxen")
```

```python
sns.catplot(data=df, x="age", y="class", hue="sex", kind="violin", bw=.25, cut=0, split=True,)
```

```python
sns.catplot( data=df, x="class", y="survived", col="sex", kind="bar", height=4, aspect=.6,)
```

```python
sns.catplot(data=df, x="age", y="class", kind="violin", color=".9", inner=None)
sns.swarmplot(data=df, x="age", y="class", size=3)
```

```python
g = sns.catplot( data=df, x="who", y="survived", col="class", kind="bar", height=4, aspect=.6,)
g.set_axis_labels("", "Survival Rate")
g.set_xticklabels(["Men", "Women", "Children"])
g.set_titles("{col_name} {col_var}")
g.set(ylim=(0, 1))
g.despine(left=True)
```
